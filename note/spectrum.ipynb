# 读取IBD谱

```python
import h5py
import numpy as np
#读取../data/hdf5/dayabay_dataset/dayabay_ibd_spectra_8AD.hdf5
data_path = '../data/hdf5/dayabay_dataset/dayabay_ibd_spectra_8AD.hdf5'
with h5py.File(data_path, 'r') as file:
    #查看file的结构
    print(file.keys())
    #读取'ibd_spectrum_AD11'
    ibd_spectrum = file['ibd_spectrum_AD11'][:]
    print(ibd_spectrum.shape)
```

```python
E_min_MeV=ibd_spectrum['E_min_MeV']
E_max_MeV=ibd_spectrum['E_max_MeV']
Counts=ibd_spectrum['N']
print(E_min_MeV)
print(E_max_MeV)
print(Counts)
```

```python
np.sum(Counts)
```
